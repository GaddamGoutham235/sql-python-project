```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/laptop-cleaned-v1.csv")
```

```python
df
```

```python
df.drop('index',axis=1,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Weight'] = df['Weight'].astype(float)
```

```python
# Distribution of Companies
plt.figure(figsize=(10, 6))
sns.countplot(y='Company', data=df, order=df['Company'].value_counts().index)
plt.title('Distribution of Companies')
plt.show()
```

```python

# Distribution of Ram
plt.figure(figsize=(10, 6))
sns.countplot(x='Ram', data=df)
plt.title('Distribution of Ram Sizes')
plt.show()
```

```python
# Distribution of Price
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True)
plt.title('Distribution of Prices')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Ram', y='Price', data=df)
plt.title('Price vs. Ram')
plt.show()
```

```python
# Pairplot for selected features
sns.pairplot(df[['Price', 'Ram', 'Inches', 'Weight']])
plt.show()
```

```python
# Visualization: Price vs. Screen Size
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Inches', y='Price', hue='Company', data=df)
plt.title('Price vs. Screen Size')
plt.show()
```

```python
# Visualization: Price vs. Weight
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Weight', y='Price', hue='Company', data=df)
plt.title('Price vs. Weight')
plt.show()
```
